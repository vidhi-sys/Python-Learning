```python
import pandas as pd
data=pd.read_csv('Airbnb_Open_Data.csv')
data.head()
```

```python
data.columns
```

```python
columns_to_keep=['id', 'NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews']
data=data[columns_to_keep]
```

Filtering Columns

```python
data.shape
```

Renaming Columns

```python
data.rename(columns={'NAME':'name',})
```

check duplicates

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().head(5)
```

Null values

```python
data.isna().sum()
```

```python
data=data.dropna()
data.isna().sum()
```

```python
print(data.columns.tolist())
```

```python
data.drop(columns=['host_identity_verified'], inplace=True)
```

```python
data['host name']=data['host name'].fillna('Unknown')
```

```python
data.head()
```

```python
data['instant_bookable'] = data['instant_bookable'].fillna(False).astype(int)
```
